```python
# <------------------------------Splitting X and Y for SPORTS car------------------------------>
sportX=dfsport.drop(columns=['ID','model','engine_power','lat','lon','price']) #dropping these values since these are not helpful in prediction
sportY=dfsport['price'] #taking price in a separate variable since we need to apply linear regression
#Linear Regression basically finds relationship between these two variables and gives us output
print(sportX.to_string())
print(sportY.to_string())
# <------------------------------Splitting X and Y for LOUNGE car------------------------------>
loungeX=dflounge.drop(columns=['ID','model','engine_power','lat','lon','price'])
loungeY=dflounge['price']
print(loungeX.to_string())
print(loungeY.to_string())
# <------------------------------Splitting X and Y for POP car------------------------------>
popX=dfpop.drop(columns=['ID','model','engine_power','lat','lon','price'])
popY=dfpop['price']
print(popX.to_string())
print(popY.to_string())
```

```python
from sklearn.linear_model import LinearRegression  #syntax for importing and using Linear Regression
```

```python
# <------------------------------Creating Varibles Named MODEL and assigning/linking the Linear Regression------------------------------>
sport_model=LinearRegression()  #creating a variable named "sports_model" and assigning the linear regression imported in the before code
lounge_model=LinearRegression() #creating a variable named "lounge_model" and assigning the linear regression imported in the before code
pop_model=LinearRegression()    #creating a variable named "pop_model" and assigning the linear regression imported in the before code
#fitting the values in the model, the values were declared above
sport_model.fit(sportX,sportY) #fitting values for sportscar
lounge_model.fit(loungeX,loungeY) #fitting values for loungecar
pop_model.fit(popX,popY) #fitting values in the popcar
```

```python
#<------------------------------Predicting the values for all the car models------------------------------>
sport_ans=sport_model.predict([[2190,70000,1]]) #We are predicting the values using predict() for SPORTS model
#Here the above bracket represents[age_in_days,km,previous_owners]
print(sport_ans)
lounge_ans=lounge_model.predict([[2190,70000,1]]) #We are predicting the values using predict() for LOUNGE model
print(lounge_ans)
pop_ans=pop_model.predict([[2190,70000,1]])  #We are predicting the values using predict() for POP model
print(pop_ans)
```

```python
newdf1=pd.get_dummies(newdf,dtype=int)
#In Linear Regression, get_dummies() is used to convert strings to integers
#In newdf1,the columns we have are 'age_in_days', 'km','pervious_owners' and 'model' (that are accssed by the newdf1 inside the brackets)
# But in model we have values as 'Sports', 'Pop' and 'Lounge' so basically it is a string.
# But in Linear Regression we should compare only between (integers and integers) not between (integers and string)
#Therefore here comes getdummies(), getdummies() is used to convert string as integers
#Here there are three unique elements as pop,lounge and sports. So three columns such as "model_sport", "model_pop" and "model_sport" are automatically created
#The zeros in "model_sport" indicates that the model is not sport and the one in "model_sport" indicates that the model is sport and so for the other models
print(newdf1.to_string())
```

```python
X1=newdf1 #We are taking the above as X1
Y1=flatdf['price'] #We need only price in Y1 to put Linear Regressison
reg=LinearRegression() #Creating variable nmaed reg and calling the LinearRegression
reg.fit(X1,Y1)
```

```python
ans=reg.predict([[2190,70000,1,0,0,1]])
print(ans)
```
