```python
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv("Top Indian Places to Visit.csv")
df
```

```python
df1 = df[df['Entrance Fee in INR'] != 0]
df1=df1.drop(columns=['Unnamed: 0','Zone','State','City','Name','Establishment Year','time needed to visit in hrs','Airport with 50km Radius','Weekly Off','Significance','DSLR Allowed','Best Time to visit'])
df1
```

```python
# col=['Establishment Year','time needed to visit in hrs','Google review rating','Number of google review in lakhs']
# newdf=df1[col].fillna(df1[col].mean(), inplace=True)
# newdf
```

```python
# col=['Google review rating','Entrance Fee in INR','Number of google review in lakhs']
# newdf=df1[col].fillna(df1[col].mean())
# newdf
col=['Type','Google review rating','Entrance Fee in INR','Number of google review in lakhs']
df1[col]=df1[col].fillna(0)
```

```python
newdf=pd.get_dummies(df1,dtype=int)
newdf
```

```python
X=newdf.drop(columns=['Entrance Fee in INR'])
Y=newdf['Entrance Fee in INR']
Y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
linreg=LinearRegression()
linreg.fit(X,Y)
```

```python
linreg.score(X_train,Y_train)
```

```python
linreg.score(X_test,Y_test)
```

```python
from sklearn.linear_model import Lasso
lasreg=Lasso(alpha=25,max_iter=100,tol=0.1)
lasreg.fit(X_train,Y_train)
```

```python
lasreg.score(X_train,Y_train)
```

```python
lasreg.score(X_test,Y_test)
```

```python
from sklearn.linear_model import Ridge
ridgereg=Ridge(alpha=25,max_iter=100,tol=0.1)
ridgereg.fit(X_train,Y_train)
```

```python
ridgereg.score(X_train,Y_train)
```

```python
ridgereg.score(X_test,Y_test)
```
