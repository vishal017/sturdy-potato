```python
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv("Melbourne_housing_FULL.csv")
df
```

```python
df1=df.drop(columns=['Address','Date','Postcode','YearBuilt','Lattitude','Longtitude'])
df1
```

```python
df1=df1.dropna()
print(df1.nunique().to_string())
```

```python
df1=pd.get_dummies(df1,dtype=int)
df1
```

```python
#df1 = df1.replace({'Abbotsford':0, 'Wollert':1,'Yarraville'},{'h':0,'u':1,'t':1})
```

```python
from sklearn.linear_model import LinearRegression
```

```python
X=df1.drop(columns=['Price'])
Y=df1['Price']
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.3)
```

```python
model=LinearRegression()
model.fit(X,Y)
```

```python
ans=model.predict(X_train)
ans
```

```python
model.score(X_train,Y_train)
```

```python
model.score(X_test,Y_test)
```

```python
from sklearn.linear_model import Lasso
lasso=Lasso(alpha=25,max_iter=100,tol=0.1)
lasso.fit(X_train,Y_train)
```

```python
lasso.score(X_train,Y_train)
```

```python
lasso.score(X_test,Y_test)
```
